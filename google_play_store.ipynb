## **Size**

```python
np.dtype(dfc['Size'])
```

```python
dfc['Size'].unique()
```

There are three Kinds of Output:

a) suffixed with M(Million)

b) suffixed with k(Thousand)

c) One record has value as "Varies with device"

```python
dfc['Size'].isnull().sum()
```

```python
dfc['Size'] = dfc['Size'].str.replace('M','000')
dfc['Size'] = dfc['Size'].str.replace('k','')
dfc['Size'] = dfc['Size'].replace('Varies with device',np.nan)
```

```python
dfc['Size']
```

```python
dfc.info()
```

```python
dfc['Size'] = dfc['Size'].astype(float)
```

```python
dfc.info()
```

### **Installs, Price**

```python
dfc[['Installs','Price']]
```

```python
chars_to_remove = ['+' , '$' , ',']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        dfc[cols]=dfc[cols].str.replace(item,'')
```

## **Which Category has largest number of installations??**

```python
df_cat_installs = dfc.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000

#converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2, palette='hls' )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

### **Insights**

1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store

## **What are the Top 5 most installed Apps in Each popular Categories ??**

```python
# Grouping by 'Category' and 'App', and summing installations
dfa = dfc.groupby(['Category', 'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending=False)

# Corrected apps list with a missing comma added
apps = ['GAME', 'COMMUNICATION', 'TOOLS', 'PRODUCTIVITY', 'SOCIAL'] 

# Setting up Seaborn context and style
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40, 30))

# Loop through each app category
for i, app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)  # Get the top 5 apps in this category
    plt.subplot(5, 2, i + 1)  # Adjusting subplot for each app
    sns.barplot(data=df3, x='Installs', y='App', palette='hls')  # Bar plot of installations
    plt.xlabel('Installation in Millions')  # X-axis label
    plt.ylabel('')  # Y-axis label (optional)
    plt.title(app, size=20)  # Title for each subplot

# Adjust layout for better spacing
plt.tight_layout()
plt.subplots_adjust(hspace=0.3)
plt.show()
```

### **Insights**
1. Most popular game is Subway Surfers.
2. Most popular communication app is Hangouts.
3. Most popular productivity app is Google Drive.
4. Most popular social app is Instagram.
